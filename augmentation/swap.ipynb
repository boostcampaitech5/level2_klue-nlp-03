## 0. 호출

```python
from matplotlib import pyplot as plt
from tqdm import tqdm
import pandas as pd
```

```python
train_data = pd.read_csv ("/opt/git/level2_klue-nlp-03/data/no_noise_train.csv")
```

```python
train_data
```

## 1. 동등관계 Label Swap

```python
aug_label = [
'per:title',
'org:alternate_names',
'per:alternate_names',
'per:other_family',
'per:colleagues',
'per:siblings',
'per:spouse'
]
```

```python
aug_rows = train_data[train_data['label'].isin(aug_label)]
```

```python
aug_rows['subject_entity'], aug_rows['object_entity'] = aug_rows['object_entity'], aug_rows['subject_entity']
```

```python
swap_augmented_data = pd.concat([train_data,aug_rows], ignore_index=True)
```

```python
swap_augmented_data.to_csv('swap_augmented_data.csv',index=False)
```

## 2. 대칭관계 Label Swap

```python
aug_label2 = [
'org:founded',
'org:founded_by',
'per:children',
'per:parents'
]
```

```python
aug_rows2 = train_data[train_data['label'].isin(aug_label2)]
```

```python
modified_rows = pd.DataFrame(columns=aug_rows2.columns)
for index, row in aug_rows2.iterrows():
    if row['label'] == 'org:founded':
        row['label'] = 'org:founded_by'
        row['subject_entity'], row['object_entity'] = row['object_entity'], row['subject_entity']
    elif row['label'] == 'org:founded_by':
        row['label'] = 'org:founded'
        row['subject_entity'], row['object_entity'] = row['object_entity'], row['subject_entity']
    elif row['label'] == 'per:children':
        row['label'] = 'per:parents'
        row['subject_entity'], row['object_entity'] = row['object_entity'], row['subject_entity']
    elif row['label'] == 'per:parents':
        row['label'] = 'per:children'
        row['subject_entity'], row['object_entity'] = row['object_entity'], row['subject_entity']
    modified_rows = pd.concat([modified_rows, row.to_frame().T])
```

```python
swap_augmented_data3 = pd.concat([train_data,modified_rows], ignore_index=True)
```

```python
swap_augmented_data3.to_csv('no_noise_swap.csv',index=False)
```

## 3. no_relation 타노스

```python
no_relation_indices = swap_augmented_data3[swap_augmented_data3['label'] == 'no_relation'].index

# "no_relation"인 절반의 샘플을 삭제
half_no_relation_indices = no_relation_indices[:len(no_relation_indices)//2]
half_no_relation = swap_augmented_data3.drop(half_no_relation_indices)
```

```python
half_no_relation.to_csv('swap_augmented_data3.csv',index=False)
```
